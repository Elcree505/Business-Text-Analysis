## Library Import

```python
from fastapi import FastAPI
from schemas.meeting import MeetingInput
from schemas.result import AnalysisResult
from services.analyzer import analyze
```

### Run

```python
app = FastAPI(title="Business Meeting Analyzer")

@app.get("/health")
def health():
    return {"status": "ok"}

@app.post("/analyze", response_model=AnalysisResult)
def analyze_one(meeting: MeetingInput):
    out = analyze(meeting.title, meeting.description or "")
    return AnalysisResult(**out)
```
